3번째 프린트 함수는 문자열 포맷팅(Formatting)한 것이다.
-> 가장 선호하는 방식

{:,} 같은 경우 숫자 3자리마다 ,를 찍어주라는 의미이다.

```python
num1 = int(input('첫 번째 정수 입력: '))
num2 = int(input('두 번째 정수 입력: '))

sum = num1 + num2

print(num1, '과(와)', num2, '의 합은', sum, '입니다.')
print(str(num1) + '과(와)'+ str(num2)+ '의 합은'+ str(sum) + '입니다.')
print('{}과(와) {}의 합은 {}입니다.'.format(num1,num2,sum))
print('{0:,}과(와) {1:,}의 합은 {2:,}입니다.'.format(num1,num2,sum))
print('%d과(와) %d의 합은 %d입니다.'%(num1,num2,sum))
```

기차요금 = 목적지 + 열차종류 + 좌석종류

```python
def dest(d):
    if d == '서울':
        a = 10000
    elif d == '대전':
        a = 8000
    elif d == '대구':
        a = 6000
    elif d == '울산':
        a = 5000
    elif d == '부산':
        a = 4000
    return a

def kinds(t):
    if t == 'ktx':
        b = 20000
    elif t == '새마을호':
        b = 15000
    elif t == '무궁화호' :
        b = 10000
    elif t == '통일호' :
        b = 5000
    return b

def seat(s):
    if s == '좌석':
        c = 5000
    else:
        c = 0
    return c

def payment(a, b, c):
    result = a + b + c
    print('총 기차요금은 {:,}원 입니다.'.format(result))
```

```python
d = input('1. 목적지 입력\n서울, 대전, 대구, 울산, 부산\n')
a = dest(d)

k = input('2. 열차종류 입력\nktx, 새마을호, 무궁화호, 통일호\n')
b = kinds(k)

s = input('3. 좌석종류 입력\n좌석, 입석\n')
c = seat(s)

payment(a, b, c)
```

음료가격 = 음료종류 + 음료크기


```python
def drinks(drink):
    if drink == '아메리카노':
        drinksprice = 5000
    elif drink == '카페라떼':
        drinksprice = 6000
    else :
        drinksprice = 9000
    return drinksprice

def size(si):
    if si == '스몰':
        sizeprice = 500
    elif si == '레귤러':
        sizeprice = 600
    else :
        sizeprice = 900 
    return sizeprice
def totalprice (a,b):
    result = a+b
    print('총 금액은 {:,}입니다.'.format(result))
```

```python
d = input('원하는 음료를 선택하세요.\n아메리카노, 카페라떼, 카푸치노\n')
s = input('원하는 사이즈를 선택하세요.\n스몰, 레귤러, 라지\n')

a = drinks(d)
b = size(s)
totalprice(a,b)
```

# 파일 입출력

파일에서 문자열을 읽고 쓰는 방법은 매우 중요하다.

웹 크롤링을 통해 수집된 다양한 데이터들을 파일에 기록하기 위해 open함수와 write 메소드에 대한 이해가 필요하다

w: write (파일 생성)

a: append (내용 추가)

r: read (파일에서 내용 불러옴)



```python
f= open('python.txt','w')
f.write('Hello, Python!\n')
f.close()
```

```python
f = open('python.txt','a')
f.write('I love Python!')
f.close()
```

```python
f = open('python.txt','r')
s = f.read()
print(s)
f.close()
```

with open() as f:

```python
lines = ['Python\n','Swift\n','Java\n','Kotlin']

with open('language.txt','w') as f:
    f.writelines(lines)
```

# 예외 처리

프로그램 작성 시 무수히 많은 오류가 발생하게 되는데 이러한 오류를 의도적으로 무시해야 하는 경우가 발생한다. 이런 경우 파이썬의 try, except를 사용하여 오류에 대한 예외 처리가 가능하다.

스크린샷 2023-01-05 오후 2.31.42.png

```python
while True:
    try:
        x = int(input('나누고자 하는 숫자를 입력: '))
        y = 100/x
        print(y)
    except:
        print('예외가 발생, 숫자를 0으로 나누기 금지')
        break
```

스크린샷 2023-01-05 오후 2.33.33.png

```python
while True:
    try:
        x = int(input('나누고자 하는 숫자를 입력: '))
        y = 100/x
    except:
        print('예외가 발생, 숫자를 0으로 나누기 금지')
        break
    else:
        print(y)
    finally:
        print('-' * 40)
```

# HTML 기본문법

웹 크롤링 / 웹 스크래핑 / 웹 스파이더링

-> 제목과 링크를 대량으로 복사하는 것

WWW - 도서관

Web Site - 특정한 한 권의 책

Web Page - 책의 한 페이지 

Home Page - 첫 페이지

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

with urlopen('https://en.wikipedia.org/wiki/Main_Page') as response:
    soup = BeautifulSoup(response,'html.parser')
    
    print(soup)

    for anchor in soup.find_all('a'):
        print(anchor.get('href','/'))
```

```python
from bs4 import BeautifulSoup
import requests

pageurl = requests.get('https://search.naver.com/search.naver?sm=tab_hty.top&where=blog&query=%ED%8C%8C%EC%9D%B4%EC%8D%AC&oquery=%ED%8C%8C%EC%9D%B4%EC%8D%AC&tqi=hJi%2Bwwp0YiRssRsntvossssss%2FZ-079244')

soup = BeautifulSoup(pageurl.content,'html.parser')

result = soup.find_all(attrs={'class':'api_txt_lines total_tit'})

for i in result:
    print(i.get_text())
    print(i.attrs['href'])
```

```python
from bs4 import BeautifulSoup
import requests
import urllib.parse
import urllib.request

pageurl1 = 'https://search.naver.com/search.naver?where=blog&sm=tab_viw.blog&query='
pageurl2 = urllib.parse.quote_plus(input('원하는 검색어 입력: '))
pageurl = pageurl1 + pageurl2

html = urllib.request.urlopen(pageurl).read()

soup = BeautifulSoup(html, 'html.parser')

result = soup.find_all(attrs = {'class':'api_txt_lines total_tit'})

for i in result:
    print(i.get_text())
    print(i.attrs['href'])
```
